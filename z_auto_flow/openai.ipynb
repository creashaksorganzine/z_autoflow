```python
import openai
import soundfile as sf
import os
from dotenv import load_dotenv

# Load the OpenAI API key from .secrets.env
load_dotenv(".secrets.env")
openai.api_key = os.getenv("OPENAI_API_KEY")

# Replace the filename with the name of your audio file
audio_filename = "5_Historical_Examples_Military_Deception.mp3"

# Load the audio file and get its sample rate
audio, sample_rate = sf.read(audio_filename)

# Set the chunk size and overlap
chunk_size = 60 # seconds
overlap = 1 # seconds

# Calculate the number of chunks
n_chunks = int((len(audio) / sample_rate) // (chunk_size - overlap)) + 1

# Initialize the transcript
transcript = ""

# Iterate over the chunks
for i in range(n_chunks):
    # Calculate the start and end positions of the chunk
    start_pos = i * (chunk_size - overlap) * sample_rate
    end_pos = min(len(audio), (i + 1) * chunk_size * sample_rate)
    
    # Get the audio chunk and convert it to a string
    audio_chunk = audio[start_pos:end_pos]
    audio_chunk_string = "\n".join([str(sample) for sample in audio_chunk])

    # Set up the transcription prompt
    prompt = f"Transcribe chunk {i+1} of audio file {audio_filename}\n\n{audio_chunk_string}\n\nTranscription:"
    
    # Set up the API request
    model_engine = "davinci" # or use another model if desired
    completions = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.7,
    )
    
    # Get the transcription from the API response
    transcription = completions.choices[0].text.strip()
    
    # Append the transcription to the transcript
    transcript += transcription + "\n"

# Save the transcript to a file
with open("transcript.txt", "w") as f:
    f.write(transcript)

# Print the transcript
print(transcript)
```
